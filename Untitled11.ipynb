```python
# ============================================================================
# CELL 2: STOCK SYMBOLS AND DATA DOWNLOAD FUNCTIONS
# ============================================================================

SYMBOLS_2017 = [
    "ABB.NS", "ACC.NS", "AIAENG.NS", "ADANIENT.NS", "ADANIPORTS.NS", 
    "ADANIPOWER.NS", "ABFRL.NS", "AJANTPHARM.NS", "ALKEM.NS", "AMARAJABAT.NS",
    "AMBUJACEM.NS", "APOLLOHOSP.NS", "APOLLOTYRE.NS", "ARVIND.NS", "ASHOKLEY.NS",
    "ASIANPAINT.NS", "AUROPHARMA.NS", "DMART.NS", "AXISBANK.NS", "BAJAJ-AUTO.NS",
    "BAJFINANCE.NS", "BAJAJFINSV.NS", "BALKRISIND.NS", "BANKBARODA.NS", 
    "BANKINDIA.NS", "BATAINDIA.NS", "BERGEPAINT.NS", "BEL.NS", "BHARATFIN.NS",
    "BHARATFORG.NS", "BHEL.NS", "BPCL.NS", "BHARTIARTL.NS", "INFRATEL.NS",
    "BIOCON.NS", "BOSCHLTD.NS", "BRITANNIA.NS", "CIPLA.NS", "COALINDIA.NS",
    "COLPAL.NS", "DIVISLAB.NS", "DRREDDY.NS", "EICHERMOT.NS", "GAIL.NS",
    "GRASIM.NS", "HCLTECH.NS", "HDFCBANK.NS", "HEROMOTOCO.NS", "HINDALCO.NS",
    "HINDUNILVR.NS", "HDFC.NS", "ITC.NS", "ICICIBANK.NS", "INDUSINDBK.NS",
    "INFY.NS", "JSWSTEEL.NS", "KOTAKBANK.NS", "LT.NS", "LUPIN.NS", "M&M.NS",
    "MARUTI.NS", "NTPC.NS", "ONGC.NS", "POWERGRID.NS", "RELIANCE.NS", 
    "SBIN.NS", "SUNPHARMA.NS", "TCS.NS", "TATAMOTORS.NS", "TATASTEEL.NS",
    "TECHM.NS", "TITAN.NS", "UPL.NS", "ULTRACEMCO.NS", "WIPRO.NS"
]

# Symbols with aliases (ticker changes)
ALIASES = {
    "MOTHERSUMI.NS": ["MOTHERSON.NS"],
    "SRTRANSFIN.NS": ["SHRIRAMFIN.NS"],
    "TATAGLOBAL.NS": ["TATACONSUM.NS"],
    "INFRATEL.NS": ["INDUSTOWER.NS"],
    "CADILAHC.NS": ["ZYDUSLIFE.NS"],
}

# Delisted stocks to always exclude
DROP_ALWAYS = set(["RCOM.NS", "DHFL.NS", "VAKRANGEE.NS"])
SYMBOLS = [s for s in SYMBOLS_2017 if s not in DROP_ALWAYS]

def normalize_yf_columns(df):
    """Remove MultiIndex from yfinance dataframes"""
    if df is None or df.empty:
        return df
    if isinstance(df.columns, pd.MultiIndex):
        df = df.copy()
        df.columns = df.columns.get_level_values(0)
    return df

def download_one(ticker):
    """Download data for a single ticker"""
    try:
        df = yf.download(ticker, start=CFG["start"], end=CFG["end"], 
                        progress=False, auto_adjust=False)
        df = normalize_yf_columns(df)
        if df is None or df.empty:
            return None
        
        need = ["Open", "High", "Low", "Close", "Volume"]
        if not all(c in df.columns for c in need):
            return None
        
        df = df[need].dropna()
        return None if df.empty else df
    except Exception as e:
        return None

def download_ohlcv(symbols, min_days):
    """Download OHLCV data for all symbols with fallback to aliases"""
    data, failed = {}, []
    
    for i, s in enumerate(symbols):
        if (i + 1) % 20 == 0:
            print(f"Progress: {i+1}/{len(symbols)}")
        
        tried = [s] + ALIASES.get(s, [])
        ok, used = None, None
        
        for t in tried:
            df = download_one(t)
            if df is not None and len(df) >= min_days:
                ok, used = df, t
                break
        
        if ok is None:
            failed.append(s)
        else:
            data[s] = ok
            if used != s:
                print(f"Alias used: {s} -> {used}")
    
    print(f"\nDownload complete: {len(data)} OK, {len(failed)} Failed")
    if failed:
        print(f"Failed symbols: {failed[:15]}")
    
    return data
```

```python
# ============================================================================
# CELL 3: MINUTE DATA LOADING (FROM KAGGLE DATASET)
# ============================================================================

def load_minute_data(data_path, symbols, start_date, end_date):
    """
    Load minute-level data from Kaggle dataset
    Assumes CSV files are named like: SYMBOL.csv
    """
    minute_data = {}
    
    if not os.path.exists(data_path):
        print(f"Warning: Minute data path not found: {data_path}")
        return minute_data
    
    for symbol in symbols:
        # Remove .NS suffix for file matching
        base_symbol = symbol.replace('.NS', '')
        file_path = os.path.join(data_path, f"{base_symbol}.csv")
        
        if not os.path.exists(file_path):
            continue
        
        try:
            df = pd.read_csv(file_path)
            
            # Standardize column names
            df.columns = df.columns.str.lower()
            if 'datetime' in df.columns:
                df['date'] = pd.to_datetime(df['datetime'])
            elif 'timestamp' in df.columns:
                df['date'] = pd.to_datetime(df['timestamp'])
            else:
                df['date'] = pd.to_datetime(df.index)
            
            # Filter date range
            df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]
            
            if len(df) > 0:
                minute_data[symbol] = df
                
        except Exception as e:
            print(f"Error loading {symbol}: {e}")
    
    print(f"Loaded minute data for {len(minute_data)} symbols")
    return minute_data

def aggregate_minute_to_daily(minute_data):
    """Convert minute data to daily OHLCV"""
    daily_data = {}
    
    for symbol, df in minute_data.items():
        # Ensure we have required columns
        df['date'] = pd.to_datetime(df['date'])
        
        daily = df.groupby(df['date'].dt.date).agg({
            'open': 'first',
            'high': 'max',
            'low': 'min',
            'close': 'last',
            'volume': 'sum'
        })
        
        daily.index = pd.to_datetime(daily.index)
        daily.columns = ['Open', 'High', 'Low', 'Close', 'Volume']
        daily_data[symbol] = daily
    
    return daily_data
```

```python
# ============================================================================
# CELL 4: FEATURE ENGINEERING
# ============================================================================

def zscore(x, win=20):
    """Calculate rolling z-score"""
    return (x - x.rolling(win).mean()) / (x.rolling(win).std() + 1e-12)

def compute_features(df):
    """Compute technical indicators and features"""
    out = pd.DataFrame(index=df.index)
    c, h, l = df["Close"], df["High"], df["Low"]
    v = df["Volume"].replace(0, np.nan).ffill()
    
    # Returns
    r1 = c.pct_change()
    out["ret1"] = r1
    out["ret5"] = c.pct_change(5)
    out["ret10"] = c.pct_change(10)
    out["mom20"] = c / c.shift(20) - 1
    
    # Volatility
    out["vol20"] = r1.rolling(20).std()
    out["vol60"] = r1.rolling(60).std()
    
    # Moving averages
    out["sma10"] = c.rolling(10).mean() / c - 1
    out["sma20"] = c.rolling(20).mean() / c - 1
    out["sma50"] = c.rolling(50).mean() / c - 1
    out["sma200"] = c.rolling(200).mean() / c - 1
    
    # True Range & ATR
    tr = (h - l) / c
    out["tr"] = tr
    out["atr14"] = tr.rolling(14).mean()
    
    # RSI
    delta = c.diff()
    up = delta.clip(lower=0)
    down = (-delta).clip(lower=0)
    rs = up.rolling(14).mean() / (down.rolling(14).mean() + 1e-12)
    out["rsi14"] = 100 - 100 / (1 + rs)
    
    # MACD
    ema12 = c.ewm(span=12, adjust=False).mean()
    ema26 = c.ewm(span=26, adjust=False).mean()
    macd = ema12 - ema26
    sig = macd.ewm(span=9, adjust=False).mean()
    out["macd"] = macd / c
    out["macd_sig"] = sig / c
    out["macd_hist"] = (macd - sig) / c
    
    # Volume features
    out["logvol_z20"] = zscore(np.log(v + 1), 20)
    out["vol_chg5"] = v.pct_change(5)
    
    return out.replace([np.inf, -np.inf], np.nan)

FEATURES = [
    "ret1", "ret5", "ret10", "mom20", "vol20", "vol60",
    "sma10", "sma20", "sma50", "sma200", "tr", "atr14",
    "rsi14", "macd", "macd_sig", "macd_hist", "logvol_z20", "vol_chg5"
]
```

```python
# ============================================================================
# CELL 5: BUILD PANEL DATA
# ============================================================================

def build_symbol_arrays(data_dict):
    """Create arrays for each symbol and build panel dataset"""
    sym_arrays = {}
    panel_rows = []
    
    for sym, df in data_dict.items():
        feat = compute_features(df)[FEATURES]
        y = (df["Close"].shift(-CFG["horizon"]) / df["Close"] - 1.0).astype(float)
        
        tmp = feat.copy()
        tmp["y"] = y
        tmp["close"] = df["Close"]
        tmp = tmp.dropna(subset=FEATURES + ["y", "close"])
        
        if len(tmp) < (CFG["lookback"] + 300):
            continue
        
        dates = pd.to_datetime(tmp.index)
        X = tmp[FEATURES].values.astype(np.float32)
        yv = tmp["y"].values.astype(np.float32)
        close = tmp["close"].values.astype(np.float32)
        
        sym_arrays[sym] = {
            "dates": dates,
            "X": X,
            "y": yv,
            "close": close
        }
        
        part = pd.DataFrame(X, columns=FEATURES)
        part["y"] = yv
        part["symbol"] = sym
        part["date"] = dates
        panel_rows.append(part)
    
    panel = pd.concat(panel_rows, ignore_index=True)
    panel = panel.sort_values(["date", "symbol"]).reset_index(drop=True)
    
    return sym_arrays, panel

def make_pairs(sym_arrays):
    """Create (symbol, time_index) pairs for LSTM training"""
    pairs = []
    for sym, d in sym_arrays.items():
        for t in range(CFG["lookback"], len(d["dates"])):
            pairs.append((sym, t))
    return pairs
```

```python
# ============================================================================
# CELL 6: LSTM DATASET AND MODEL
# ============================================================================

class SeqSliceDataset(Dataset):
    def __init__(self, sym_arrays, pairs, start_date=None, end_date=None, standardize=None):
        self.sym_arrays = sym_arrays
        sd = pd.to_datetime(start_date) if start_date else None
        ed = pd.to_datetime(end_date) if end_date else None
        
        kept = []
        for sym, t in pairs:
            d = sym_arrays[sym]["dates"][t]
            if sd is not None and d < sd:
                continue
            if ed is not None and d > ed:
                continue
            kept.append((sym, t))
        
        self.pairs = kept
        self.mu, self.sig = (standardize if standardize else (None, None))
    
    def __len__(self):
        return len(self.pairs)
    
    def __getitem__(self, i):
        sym, t = self.pairs[i]
        arr = self.sym_arrays[sym]
        X = arr["X"][t - CFG["lookback"]:t]
        y = arr["y"][t]
        
        if self.mu is not None:
            X = (X - self.mu) / self.sig
        
        date_int = int(arr["dates"][t].value)
        return (
            torch.from_numpy(X),
            torch.tensor([y], dtype=torch.float32),
            sym,
            torch.tensor(date_int, dtype=torch.int64)
        )

class LSTMRegressor(nn.Module):
    def __init__(self, n_features, hidden=64, dropout=0.3):
        super().__init__()
        self.ln = nn.LayerNorm(n_features)
        self.lstm = nn.LSTM(n_features, hidden, batch_first=True, num_layers=2)
        self.drop = nn.Dropout(dropout)
        self.fc = nn.Sequential(
            nn.Linear(hidden, 32),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(32, 1)
        )
    
    def forward(self, x):
        x = self.ln(x)
        x, _ = self.lstm(x)
        x = self.drop(x)
        return self.fc(x[:, -1, :])

def train_lstm(train_loader, val_loader, n_features):
    """Train LSTM model with early stopping"""
    device = CFG["device"]
    model = LSTMRegressor(
        n_features,
        hidden=CFG["lstm_hidden"],
        dropout=CFG["lstm_dropout"]
    ).to(device)
    
    opt = torch.optim.Adam(model.parameters(), lr=CFG["lstm_lr"])
    loss_fn = nn.MSELoss()
    
    best = float("inf")
    best_state = None
    patience = 3
    bad = 0
    
    for ep in range(1, CFG["lstm_epochs"] + 1):
        # Training
        model.train()
        tl = []
        for xb, yb, _, _ in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            opt.zero_grad(set_to_none=True)
            pred = model(xb)
            loss = loss_fn(pred, yb)
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            opt.step()
            tl.append(loss.item())
        
        # Validation
        model.eval()
        vl = []
        with torch.no_grad():
            for xb, yb, _, _ in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                vl.append(loss_fn(model(xb), yb).item())
        
        v = float(np.mean(vl)) if vl else float("inf")
        print(f"Epoch {ep:02d} | Train MSE: {np.mean(tl):.6f} | Val MSE: {v:.6f}")
        
        # Early stopping
        if v < best - 1e-6:
            best = v
            best_state = {k: vv.detach().cpu().clone() 
                         for k, vv in model.state_dict().items()}
            bad = 0
        else:
            bad += 1
            if bad >= patience:
                print(f"Early stopping at epoch {ep}")
                break
    
    if best_state is not None:
        model.load_state_dict(best_state)
    
    return model

@torch.no_grad()
def predict_lstm(model, loader):
    """Generate LSTM predictions"""
    device = CFG["device"]
    model.eval()
    out = []
    
    for xb, _, syms, date_int in loader:
        xb = xb.to(device)
        pred = model(xb).detach().cpu().numpy().reshape(-1)
        dates = pd.to_datetime(date_int.numpy(), unit="ns")
        out.extend(list(zip(dates, list(syms), pred)))
    
    df = pd.DataFrame(out, columns=["date", "symbol", "pred_lstm"])
    df["date"] = pd.to_datetime(df["date"])
    return df
```

```python
# ============================================================================
# CELL 7: EVALUATION METRICS
# ============================================================================

def calculate_metrics(y_true, y_pred, prefix=""):
    """Calculate regression metrics"""
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    
    # Direction accuracy
    direction_acc = np.mean(np.sign(y_true) == np.sign(y_pred))
    
    metrics = {
        f"{prefix}MSE": mse,
        f"{prefix}RMSE": rmse,
        f"{prefix}MAE": mae,
        f"{prefix}R2": r2,
        f"{prefix}Direction_Acc": direction_acc
    }
    
    return metrics

def sharpe(returns, ann=252):
    """Calculate Sharpe ratio"""
    r = np.asarray(returns, float)
    if len(r) < 2 or r.std() == 0:
        return 0.0
    return (r.mean() / r.std()) * math.sqrt(ann)

def max_drawdown(equity):
    """Calculate maximum drawdown"""
    e = np.asarray(equity, float)
    peak = np.maximum.accumulate(e)
    dd = (e / peak) - 1
    return float(dd.min())

def cagr(equity, ann=252):
    """Calculate Compound Annual Growth Rate"""
    e = np.asarray(equity, float)
    if len(e) < 2:
        return 0.0
    years = (len(e) - 1) / ann
    if years <= 0:
        return 0.0
    return float((e[-1] / e[0]) ** (1 / years) - 1)

def calmar_ratio(equity, ann=252):
    """Calculate Calmar ratio (CAGR / Max Drawdown)"""
    c = cagr(equity, ann)
    mdd = abs(max_drawdown(equity))
    return c / mdd if mdd > 0 else 0.0

def sortino_ratio(returns, ann=252):
    """Calculate Sortino ratio"""
    r = np.asarray(returns, float)
    if len(r) < 2:
        return 0.0
    downside = r[r < 0].std()
    if downside == 0:
        return 0.0
    return (r.mean() / downside) * math.sqrt(ann)
```

```python
# ============================================================================
# CELL 8: ENHANCED BACKTESTING ENGINE (REPLACE ORIGINAL)
# ============================================================================

def get_rebalance_dates(dates, weekday=4):
    """Get rebalancing dates (e.g., every Friday)"""
    idx = pd.DatetimeIndex(sorted(pd.to_datetime(dates).unique()))
    return idx[idx.weekday == weekday]

def run_backtest_enhanced(panel_pred, sym_arrays, idx_close, start_date, end_date, name="Backtest"):
    """
    Enhanced backtest with detailed buy/sell logging
    """
    panel_pred = panel_pred.copy()
    panel_pred["date"] = pd.to_datetime(panel_pred["date"])
    
    start = pd.to_datetime(start_date)
    end = pd.to_datetime(end_date)
    
    dates = pd.DatetimeIndex(sorted(panel_pred["date"].unique()))
    dates = dates[(dates >= start) & (dates <= end)]
    
    if len(dates) == 0:
        print(f"Warning: No dates in range {start} to {end}")
        return None
    
    rebal = set(get_rebalance_dates(dates, CFG["rebalance_weekday"]))
    
    equity = [1.0]
    eq_dates = [dates[0]]
    daily = []
    pos = {}
    trades = []
    holdings_history = []
    detailed_transactions = []  # NEW: Detailed buy/sell log
    
    for i in range(1, len(dates)):
        d = dates[i]
        prev = dates[i - 1]
        
        # Calculate daily return
        day = 0.0
        for sym, w in pos.items():
            arr = sym_arrays.get(sym)
            if arr is None:
                continue
            
            dt = arr["dates"].values
            t0 = np.searchsorted(dt, np.datetime64(prev))
            t1 = np.searchsorted(dt, np.datetime64(d))
            
            if t0 >= len(arr["close"]) or t1 >= len(arr["close"]):
                continue
            if dt[t0] != np.datetime64(prev) or dt[t1] != np.datetime64(d):
                continue
            
            day_return = float(arr["close"][t1] / arr["close"][t0] - 1.0)
            day += w * day_return
        
        equity.append(equity[-1] * (1 + day))
        eq_dates.append(d)
        daily.append(day)
        
        # Rebalancing logic with detailed logging
        if d in rebal:
            today = panel_pred[panel_pred["date"] == d].dropna(subset=["score"])
            today = today.sort_values("score", ascending=False)
            
            if len(today) >= CFG["top_k"]:
                picks = today["symbol"].head(CFG["top_k"]).tolist()
                w = 1.0 / len(picks)
                new_pos = {s: w for s in picks}
                
                # Identify buys and sells
                old_stocks = set(pos.keys())
                new_stocks = set(new_pos.keys())
                
                sells = old_stocks - new_stocks
                buys = new_stocks - old_stocks
                holds = old_stocks & new_stocks
                
                # Log detailed transactions
                portfolio_value = equity[-1]
                
                for stock in sells:
                    detailed_transactions.append({
                        'date': d,
                        'action': 'SELL',
                        'symbol': stock,
                        'weight': pos[stock],
                        'value': portfolio_value * pos[stock],
                        'portfolio_value': portfolio_value
                    })
                
                for stock in buys:
                    detailed_transactions.append({
                        'date': d,
                        'action': 'BUY',
                        'symbol': stock,
                        'weight': new_pos[stock],
                        'value': portfolio_value * new_pos[stock],
                        'portfolio_value': portfolio_value
                    })
                
                # Calculate turnover and costs
                union = set(pos) | set(new_pos)
                turnover = sum(abs(new_pos.get(s, 0.0) - pos.get(s, 0.0)) for s in union)
                cost = (CFG["cost_bps"] + CFG["slip_bps"]) * 1e-4 * turnover
                equity[-1] *= (1 - cost)
                
                # Record trade
                trades.append({
                    "date": d,
                    "action": "rebalance",
                    "turnover": turnover,
                    "cost": cost,
                    "holdings": picks,
                    "buys": list(buys),
                    "sells": list(sells),
                    "holds": list(holds),
                    "num_buys": len(buys),
                    "num_sells": len(sells)
                })
                
                pos = new_pos
                holdings_history.append({"date": d, "holdings": picks})
    
    # Calculate benchmark (index) returns
    idx_close = idx_close.dropna()
    idx_ret = idx_close.reindex(eq_dates).pct_change().fillna(0.0).values
    idx_eq = [1.0]
    for r in idx_ret[1:]:
        idx_eq.append(idx_eq[-1] * (1 + float(r)))
    
    # Calculate statistics
    stats = {
        "Period": name,
        "Start": eq_dates[0].strftime("%Y-%m-%d"),
        "End": eq_dates[-1].strftime("%Y-%m-%d"),
        "Total_Return": (equity[-1] - 1.0) * 100,
        "CAGR": cagr(equity) * 100,
        "Sharpe": sharpe(daily),
        "Sortino": sortino_ratio(daily),
        "MaxDD": max_drawdown(equity) * 100,
        "Calmar": calmar_ratio(equity),
        "Num_Trades": len(trades),
        "Avg_Turnover": np.mean([t["turnover"] for t in trades]) if trades else 0,
        "Total_Cost": sum([t["cost"] for t in trades]) * equity[0],
        "Best_Day": max(daily) * 100 if daily else 0,
        "Worst_Day": min(daily) * 100 if daily else 0,
        "Win_Rate": np.mean([d > 0 for d in daily]) * 100 if daily else 0,
        "Benchmark_Return": (idx_eq[-1] - 1.0) * 100,
        "Benchmark_CAGR": cagr(idx_eq) * 100,
        "Benchmark_Sharpe": sharpe(idx_ret[1:]),
        "Benchmark_MaxDD": max_drawdown(idx_eq) * 100,
        "Alpha": (equity[-1] - idx_eq[-1]) * 100,
        "Final_Portfolio_Value": equity[-1],
        "Total_Buys": sum([t["num_buys"] for t in trades]),
        "Total_Sells": sum([t["num_sells"] for t in trades])
    }
    
    return {
        "dates": eq_dates,
        "equity": equity,
        "idx_equity": idx_eq,
        "daily_returns": daily,
        "stats": stats,
        "trades": trades,
        "holdings_history": holdings_history,
        "detailed_transactions": detailed_transactions  # NEW
    }
```

```python


# ============================================================================
# CELL 8: ENHANCED BACKTESTING ENGINE (REPLACE ORIGINAL)
# ============================================================================

def get_rebalance_dates(dates, weekday=4):
    """Get rebalancing dates (e.g., every Friday)"""
    idx = pd.DatetimeIndex(sorted(pd.to_datetime(dates).unique()))
    return idx[idx.weekday == weekday]

def run_backtest_enhanced(panel_pred, sym_arrays, idx_close, start_date, end_date, name="Backtest"):
    """
    Enhanced backtest with detailed buy/sell logging
    """
    panel_pred = panel_pred.copy()
    panel_pred["date"] = pd.to_datetime(panel_pred["date"])
    
    start = pd.to_datetime(start_date)
    end = pd.to_datetime(end_date)
    
    dates = pd.DatetimeIndex(sorted(panel_pred["date"].unique()))
    dates = dates[(dates >= start) & (dates <= end)]
    
    if len(dates) == 0:
        print(f"Warning: No dates in range {start} to {end}")
        return None
    
    rebal = set(get_rebalance_dates(dates, CFG["rebalance_weekday"]))
    
    equity = [1.0]
    eq_dates = [dates[0]]
    daily = []
    pos = {}
    trades = []
    holdings_history = []
    detailed_transactions = []  # NEW: Detailed buy/sell log
    
    for i in range(1, len(dates)):
        d = dates[i]
        prev = dates[i - 1]
        
        # Calculate daily return
        day = 0.0
        for sym, w in pos.items():
            arr = sym_arrays.get(sym)
            if arr is None:
                continue
            
            dt = arr["dates"].values
            t0 = np.searchsorted(dt, np.datetime64(prev))
            t1 = np.searchsorted(dt, np.datetime64(d))
            
            if t0 >= len(arr["close"]) or t1 >= len(arr["close"]):
                continue
            if dt[t0] != np.datetime64(prev) or dt[t1] != np.datetime64(d):
                continue
            
            day_return = float(arr["close"][t1] / arr["close"][t0] - 1.0)
            day += w * day_return
        
        equity.append(equity[-1] * (1 + day))
        eq_dates.append(d)
        daily.append(day)
        
        # Rebalancing logic with detailed logging
        if d in rebal:
            today = panel_pred[panel_pred["date"] == d].dropna(subset=["score"])
            today = today.sort_values("score", ascending=False)
            
            if len(today) >= CFG["top_k"]:
                picks = today["symbol"].head(CFG["top_k"]).tolist()
                w = 1.0 / len(picks)
                new_pos = {s: w for s in picks}
                
                # Identify buys and sells
                old_stocks = set(pos.keys())
                new_stocks = set(new_pos.keys())
                
                sells = old_stocks - new_stocks
                buys = new_stocks - old_stocks
                holds = old_stocks & new_stocks
                
                # Log detailed transactions
                portfolio_value = equity[-1]
                
                for stock in sells:
                    detailed_transactions.append({
                        'date': d,
                        'action': 'SELL',
                        'symbol': stock,
                        'weight': pos[stock],
                        'value': portfolio_value * pos[stock],
                        'portfolio_value': portfolio_value
                    })
                
                for stock in buys:
                    detailed_transactions.append({
                        'date': d,
                        'action': 'BUY',
                        'symbol': stock,
                        'weight': new_pos[stock],
                        'value': portfolio_value * new_pos[stock],
                        'portfolio_value': portfolio_value
                    })
                
                # Calculate turnover and costs
                union = set(pos) | set(new_pos)
                turnover = sum(abs(new_pos.get(s, 0.0) - pos.get(s, 0.0)) for s in union)
                cost = (CFG["cost_bps"] + CFG["slip_bps"]) * 1e-4 * turnover
                equity[-1] *= (1 - cost)
                
                # Record trade
                trades.append({
                    "date": d,
                    "action": "rebalance",
                    "turnover": turnover,
                    "cost": cost,
                    "holdings": picks,
                    "buys": list(buys),
                    "sells": list(sells),
                    "holds": list(holds),
                    "num_buys": len(buys),
                    "num_sells": len(sells)
                })
                
                pos = new_pos
                holdings_history.append({"date": d, "holdings": picks})
    
    # Calculate benchmark (index) returns
    idx_close = idx_close.dropna()
    idx_ret = idx_close.reindex(eq_dates).pct_change().fillna(0.0).values
    idx_eq = [1.0]
    for r in idx_ret[1:]:
        idx_eq.append(idx_eq[-1] * (1 + float(r)))
    
    # Calculate statistics
    stats = {
        "Period": name,
        "Start": eq_dates[0].strftime("%Y-%m-%d"),
        "End": eq_dates[-1].strftime("%Y-%m-%d"),
        "Total_Return": (equity[-1] - 1.0) * 100,
        "CAGR": cagr(equity) * 100,
        "Sharpe": sharpe(daily),
        "Sortino": sortino_ratio(daily),
        "MaxDD": max_drawdown(equity) * 100,
        "Calmar": calmar_ratio(equity),
        "Num_Trades": len(trades),
        "Avg_Turnover": np.mean([t["turnover"] for t in trades]) if trades else 0,
        "Total_Cost": sum([t["cost"] for t in trades]) * equity[0],
        "Best_Day": max(daily) * 100 if daily else 0,
        "Worst_Day": min(daily) * 100 if daily else 0,
        "Win_Rate": np.mean([d > 0 for d in daily]) * 100 if daily else 0,
        "Benchmark_Return": (idx_eq[-1] - 1.0) * 100,
        "Benchmark_CAGR": cagr(idx_eq) * 100,
        "Benchmark_Sharpe": sharpe(idx_ret[1:]),
        "Benchmark_MaxDD": max_drawdown(idx_eq) * 100,
        "Alpha": (equity[-1] - idx_eq[-1]) * 100,
        "Final_Portfolio_Value": equity[-1],
        "Total_Buys": sum([t["num_buys"] for t in trades]),
        "Total_Sells": sum([t["num_sells"] for t in trades])
    }
    
    return {
        "dates": eq_dates,
        "equity": equity,
        "idx_equity": idx_eq,
        "daily_returns": daily,
        "stats": stats,
        "trades": trades,
        "holdings_history": holdings_history,
        "detailed_transactions": detailed_transactions  # NEW
    }


# ============================================================================
# CELL 9: FIXED VISUALIZATION FUNCTIONS (REPLACE ORIGINAL)
# ============================================================================

def plot_equity_curves(results_dict, title="Portfolio Performance"):
    """Plot equity curves for multiple periods"""
    valid_results = {k: v for k, v in results_dict.items() if v is not None}
    if not valid_results:
        print("No valid results to plot")
        return
    
    n_plots = len(valid_results)
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    axes = axes.flatten()
    
    for idx, (name, res) in enumerate(valid_results.items()):
        if idx >= 4:
            break
        
        ax = axes[idx]
        ax.plot(res["dates"], res["equity"], label="Strategy", linewidth=2, color='#2E86AB')
        ax.plot(res["dates"], res["idx_equity"], label="NIFTY 50", linewidth=2, 
                color='#A23B72', linestyle='--')
        
        ax.set_title(f"{name} ({res['stats']['Start']} to {res['stats']['End']})", 
                    fontsize=12, fontweight='bold')
        ax.set_xlabel("Date")
        ax.set_ylabel("Portfolio Value")
        ax.legend(loc='upper left')
        ax.grid(True, alpha=0.3)
        
        # Add performance text
        text = f"Return: {res['stats']['Total_Return']:.1f}%\n"
        text += f"Sharpe: {res['stats']['Sharpe']:.2f}\n"
        text += f"MaxDD: {res['stats']['MaxDD']:.1f}%"
        ax.text(0.02, 0.98, text, transform=ax.transAxes, 
               verticalalignment='top', bbox=dict(boxstyle='round', 
               facecolor='wheat', alpha=0.5), fontsize=9)
    
    # Hide unused subplots
    for idx in range(len(valid_results), 4):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.savefig('equity_curves.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_monthly_returns(results_dict):
    """Plot monthly returns heatmap - FIXED VERSION"""
    valid_results = {k: v for k, v in results_dict.items() if v is not None}
    if not valid_results:
        print("No valid results to plot")
        return
    
    n_plots = len(valid_results)
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    axes = axes.flatten()
    
    for idx, (name, res) in enumerate(valid_results.items()):
        if idx >= 4:
            break
        
        # Calculate monthly returns
        equity_series = pd.Series(res["equity"], index=pd.DatetimeIndex(res["dates"]))
        monthly = equity_series.resample('M').last().pct_change().dropna() * 100
        
        if len(monthly) == 0:
            axes[idx].text(0.5, 0.5, f'{name}\nInsufficient data for monthly returns', 
                          ha='center', va='center', transform=axes[idx].transAxes)
            continue
        
        # Create year-month pivot
        df_monthly = pd.DataFrame({
            'Year': monthly.index.year,
            'Month': monthly.index.month,
            'Return': monthly.values
        })
        pivot = df_monthly.pivot(index='Month', columns='Year', values='Return')
        
        # Only plot if we have data
        if not pivot.empty and pivot.shape[0] > 0 and pivot.shape[1] > 0:
            ax = axes[idx]
            sns.heatmap(pivot, annot=True, fmt='.1f', cmap='RdYlGn', center=0,
                       ax=ax, cbar_kws={'label': 'Return (%)'}, linewidths=0.5)
            ax.set_title(f"{name} - Monthly Returns (%)", fontsize=12, fontweight='bold')
            ax.set_ylabel("Month")
            ax.set_xlabel("Year")
        else:
            axes[idx].text(0.5, 0.5, f'{name}\nInsufficient data', 
                          ha='center', va='center', transform=axes[idx].transAxes)
    
    # Hide unused subplots
    for idx in range(len(valid_results), 4):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.savefig('monthly_returns_heatmap.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_drawdown(results_dict):
    """Plot drawdown curves"""
    valid_results = {k: v for k, v in results_dict.items() if v is not None}
    if not valid_results:
        print("No valid results to plot")
        return
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    axes = axes.flatten()
    
    for idx, (name, res) in enumerate(valid_results.items()):
        if idx >= 4:
            break
        
        equity = np.array(res["equity"])
        peak = np.maximum.accumulate(equity)
        drawdown = (equity / peak - 1) * 100
        
        ax = axes[idx]
        ax.fill_between(range(len(drawdown)), drawdown, 0, alpha=0.3, color='red')
        ax.plot(drawdown, color='darkred', linewidth=1.5)
        ax.set_title(f"{name} - Drawdown", fontsize=12, fontweight='bold')
        ax.set_xlabel("Days")
        ax.set_ylabel("Drawdown (%)")
        ax.grid(True, alpha=0.3)
        ax.axhline(y=res['stats']['MaxDD'], color='black', linestyle='--', 
                  label=f"Max DD: {res['stats']['MaxDD']:.1f}%")
        ax.legend()
    
    # Hide unused subplots
    for idx in range(len(valid_results), 4):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.savefig('drawdown_curves.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_returns_distribution(results_dict):
    """Plot daily returns distribution"""
    valid_results = {k: v for k, v in results_dict.items() if v is not None}
    if not valid_results:
        print("No valid results to plot")
        return
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    axes = axes.flatten()
    
    for idx, (name, res) in enumerate(valid_results.items()):
        if idx >= 4:
            break
        
        returns = np.array(res["daily_returns"]) * 100
        
        ax = axes[idx]
        ax.hist(returns, bins=50, alpha=0.7, color='steelblue', edgecolor='black')
        ax.axvline(returns.mean(), color='red', linestyle='--', 
                  linewidth=2, label=f'Mean: {returns.mean():.2f}%')
        ax.axvline(0, color='black', linestyle='-', linewidth=1)
        ax.set_title(f"{name} - Daily Returns Distribution", fontsize=12, fontweight='bold')
        ax.set_xlabel("Daily Return (%)")
        ax.set_ylabel("Frequency")
        ax.legend()
        ax.grid(True, alpha=0.3)
        
        # Add statistics
        text = f"Std: {returns.std():.2f}%\n"
        text += f"Skew: {pd.Series(returns).skew():.2f}\n"
        text += f"Kurt: {pd.Series(returns).kurtosis():.2f}"
        ax.text(0.02, 0.98, text, transform=ax.transAxes,
               verticalalignment='top', bbox=dict(boxstyle='round',
               facecolor='wheat', alpha=0.5), fontsize=9)
    
    # Hide unused subplots
    for idx in range(len(valid_results), 4):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.savefig('returns_distribution.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_rolling_metrics(results_dict, window=60):
    """Plot rolling Sharpe and volatility"""
    valid_results = {k: v for k, v in results_dict.items() if v is not None}
    if not valid_results:
        print("No valid results to plot")
        return
    
    n_results = len(valid_results)
    fig, axes = plt.subplots(n_results, 2, figsize=(16, 4*n_results))
    
    if n_results == 1:
        axes = axes.reshape(1, -1)
    
    for idx, (name, res) in enumerate(valid_results.items()):
        returns = pd.Series(res["daily_returns"], index=pd.DatetimeIndex(res["dates"][1:]))
        
        # Rolling Sharpe
        rolling_sharpe = returns.rolling(window).mean() / returns.rolling(window).std() * np.sqrt(252)
        axes[idx, 0].plot(rolling_sharpe.index, rolling_sharpe.values, linewidth=1.5)
        axes[idx, 0].set_title(f"{name} - Rolling {window}-Day Sharpe Ratio", fontweight='bold')
        axes[idx, 0].axhline(y=0, color='black', linestyle='--', alpha=0.3)
        axes[idx, 0].grid(True, alpha=0.3)
        axes[idx, 0].set_ylabel("Sharpe Ratio")
        
        # Rolling Volatility
        rolling_vol = returns.rolling(window).std() * np.sqrt(252) * 100
        axes[idx, 1].plot(rolling_vol.index, rolling_vol.values, linewidth=1.5, color='orange')
        axes[idx, 1].set_title(f"{name} - Rolling {window}-Day Volatility", fontweight='bold')
        axes[idx, 1].grid(True, alpha=0.3)
        axes[idx, 1].set_ylabel("Annualized Volatility (%)")
    
    plt.tight_layout()
    plt.savefig('rolling_metrics.png', dpi=300, bbox_inches='tight')
    plt.show()

```

```python
# ============================================================================
# CELL 12: MODEL PERFORMANCE ANALYSIS
# ============================================================================

def evaluate_model_predictions(panel, sym_arrays, periods):
    """Evaluate model prediction accuracy across different periods"""
    results = {}
    
    for period_name, (start, end) in periods.items():
        period_data = panel[(panel['date'] >= start) & (panel['date'] <= end)]
        
        if len(period_data) == 0:
            continue
        
        # Remove NaN values
        valid = period_data.dropna(subset=['y', 'pred_xgb', 'pred_lstm', 'score'])
        
        if len(valid) == 0:
            continue
        
        # Calculate metrics for each model
        xgb_metrics = calculate_metrics(valid['y'], valid['pred_xgb'], 'XGB_')
        lstm_metrics = calculate_metrics(valid['y'], valid['pred_lstm'], 'LSTM_')
        ensemble_metrics = calculate_metrics(valid['y'], valid['score'], 'Ensemble_')
        
        results[period_name] = {**xgb_metrics, **lstm_metrics, **ensemble_metrics}
    
    df = pd.DataFrame(results).T
    
    print("\n" + "="*100)
    print("MODEL PREDICTION PERFORMANCE")
    print("="*100)
    print(df.to_string())
    print("="*100)
    
    # Save to CSV
    df.to_csv('model_performance.csv')
    
    return df

def plot_prediction_scatter(panel, period_name, start_date, end_date):
    """Plot actual vs predicted returns"""
    period_data = panel[(panel['date'] >= start_date) & (panel['date'] <= end_date)]
    valid = period_data.dropna(subset=['y', 'score']).sample(min(5000, len(period_data)))
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    models = [('pred_xgb', 'XGBoost'), ('pred_lstm', 'LSTM'), ('score', 'Ensemble')]
    
    for idx, (pred_col, model_name) in enumerate(models):
        ax = axes[idx]
        
        ax.scatter(valid['y'], valid[pred_col], alpha=0.3, s=10)
        ax.plot([-0.2, 0.2], [-0.2, 0.2], 'r--', linewidth=2, label='Perfect Prediction')
        
        # Calculate R
        r2 = r2_score(valid['y'], valid[pred_col])
        
        ax.set_xlabel('Actual Return')
        ax.set_ylabel('Predicted Return')
        ax.set_title(f'{model_name} - {period_name}\nR = {r2:.4f}', fontweight='bold')
        ax.legend()
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'prediction_scatter_{period_name}.png', dpi=300, bbox_inches='tight')
    plt.show()

def analyze_feature_importance(xgb_model):
    """Analyze and plot feature importance"""
    importance = pd.DataFrame({
        'Feature': FEATURES,
        'Importance': xgb_model.feature_importances_
    }).sort_values('Importance', ascending=False)
    
    print("\n" + "="*60)
    print("FEATURE IMPORTANCE (XGBoost)")
    print("="*60)
    print(importance.to_string(index=False))
    print("="*60)
    
    # Plot
    plt.figure(figsize=(10, 8))
    plt.barh(importance['Feature'], importance['Importance'])
    plt.xlabel('Importance')
    plt.title('Feature Importance (XGBoost)', fontweight='bold', fontsize=14)
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.savefig('feature_importance.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return importance
```

```python
# ============================================================================
# CELL 13: UPDATED MAIN PIPELINE (REPLACE ORIGINAL)
# ============================================================================

def main_pipeline():
    """Main execution pipeline for the entire trading system"""
    
    print("\n" + "="*80)
    print("STOCK TRADING SYSTEM - FULL PIPELINE EXECUTION")
    print("="*80)
    
    # Step 1: Download/Load Data
    print("\n[1/10] Loading index data...")
    idx_df = yf.download(CFG["index"], start=CFG["start"], end=CFG["end"], 
                        progress=False, auto_adjust=False)
    idx_df = normalize_yf_columns(idx_df)
    idx_close = idx_df["Close"].dropna()
    idx_close.index = pd.to_datetime(idx_close.index)
    
    print("[2/10] Loading stock data...")
    # Try to load minute data first
    if os.path.exists(CFG.get("minute_data_path", "")):
        print("  Using minute-level data from Kaggle...")
        minute_data = load_minute_data(CFG["minute_data_path"], SYMBOLS, 
                                      CFG["start"], CFG["end"])
        data = aggregate_minute_to_daily(minute_data)
    else:
        print("  Downloading daily data from Yahoo Finance...")
        data = download_ohlcv(SYMBOLS, CFG["min_days"])
    
    if len(data) < 30:
        raise RuntimeError("Too few symbols. Adjust min_days or check data source.")
    
    # Step 2: Build Features
    print("[3/10] Building features and panel data...")
    sym_arrays, panel = build_symbol_arrays(data)
    print(f"  Symbols: {len(sym_arrays)} | Panel rows: {len(panel):,}")
    
    # Step 3: Train XGBoost
    print("[4/10] Training XGBoost...")
    train_end = pd.to_datetime(CFG["train_end"])
    tr = panel[panel["date"] <= train_end].dropna(subset=FEATURES + ["y"])
    print(f"  Training samples: {len(tr):,}")
    
    xgb = XGBRegressor(
        n_estimators=1500, learning_rate=0.02, max_depth=6,
        subsample=0.8, colsample_bytree=0.8, reg_lambda=2.0,
        objective="reg:squarederror", random_state=CFG["seed"], 
        n_jobs=-1, verbosity=0
    )
    xgb.fit(tr[FEATURES].values, tr["y"].values)
    panel["pred_xgb"] = xgb.predict(panel[FEATURES].values)
    
    # Step 4: Train LSTM
    print("[5/10] Training LSTM...")
    trX = panel.loc[panel["date"] <= train_end, FEATURES].values.astype(np.float32)
    mu = trX.mean(axis=0, keepdims=True)
    sig = trX.std(axis=0, keepdims=True) + 1e-6
    
    pairs = make_pairs(sym_arrays)
    
    train_ds = SeqSliceDataset(sym_arrays, pairs, start_date="2010-01-01", 
                              end_date=CFG["train_end"], standardize=(mu, sig))
    val_ds = SeqSliceDataset(sym_arrays, pairs, start_date=CFG["val_start"],
                            end_date=CFG["val_end"], standardize=(mu, sig))
    
    train_loader = DataLoader(train_ds, batch_size=CFG["lstm_batch"], 
                             shuffle=True, num_workers=0, drop_last=True)
    val_loader = DataLoader(val_ds, batch_size=CFG["lstm_batch"], 
                           shuffle=False, num_workers=0)
    
    print(f"  Training samples: {len(train_ds):,} | Validation: {len(val_ds):,}")
    model = train_lstm(train_loader, val_loader, n_features=len(FEATURES))
    
    # Step 5: Generate LSTM predictions
    print("[6/10] Generating LSTM predictions...")
    pred_ds = SeqSliceDataset(sym_arrays, pairs, start_date="2010-01-01",
                             end_date=CFG["paper_end"], standardize=(mu, sig))
    pred_loader = DataLoader(pred_ds, batch_size=1024, shuffle=False, num_workers=0)
    pred_lstm_df = predict_lstm(model, pred_loader)
    
    # Step 6: Create ensemble
    print("[7/10] Creating ensemble predictions...")
    panel2 = panel.merge(pred_lstm_df, on=["date", "symbol"], how="left")
    panel2["pred_lstm"] = panel2["pred_lstm"].fillna(panel2["pred_xgb"])
    panel2["score"] = CFG["w_xgb"] * panel2["pred_xgb"] + CFG["w_lstm"] * panel2["pred_lstm"]
    
    # Step 7: Run Backtests (USING ENHANCED VERSION)
    print("[8/10] Running backtests...")
    
    results = {}
    
    # Validation
    print("  Running validation backtest (2022-2023)...")
    results['Validation'] = run_backtest_enhanced(panel2, sym_arrays, idx_close,
                                        CFG["val_start"], CFG["val_end"], 
                                        "Validation")
    
    # Testing
    print("  Running test backtest (2024)...")
    results['Testing'] = run_backtest_enhanced(panel2, sym_arrays, idx_close,
                                     CFG["test_start"], CFG["test_end"], 
                                     "Testing")
    
    # Paper Trading
    print("  Running paper trading simulation (2025)...")
    results['Paper_Trading'] = run_backtest_enhanced(panel2, sym_arrays, idx_close,
                                           CFG["paper_start"], CFG["paper_end"],
                                           "Paper_Trading")
    
    # Step 8: Analyze Results
    print("[9/10] Analyzing results...")
    perf_summary = create_performance_summary(results)
    trade_analysis = analyze_trades(results)
    analyze_holdings(results)
    
    # Model performance
    periods = {
        'Training': (CFG["start"], CFG["train_end"]),
        'Validation': (CFG["val_start"], CFG["val_end"]),
        'Testing': (CFG["test_start"], CFG["test_end"]),
        'Paper': (CFG["paper_start"], CFG["paper_end"])
    }
    model_perf = evaluate_model_predictions(panel2, sym_arrays, periods)
    
    # Feature importance
    feat_imp = analyze_feature_importance(xgb)
    
    # Step 9: Generate Visualizations
    print("[10/10] Generating visualizations...")
    plot_equity_curves(results)
    plot_monthly_returns(results)
    plot_drawdown(results)
    plot_returns_distribution(results)
    plot_rolling_metrics(results)
    plot_metric_comparison(results)
    
    for period_name, (start, end) in periods.items():
        if period_name != 'Training':
            plot_prediction_scatter(panel2, period_name, start, end)
    
    print("\n" + "="*80)
    print("PIPELINE COMPLETE!")
    print("="*80)
    print("\nGenerated files:")
    print("  - equity_curves.png")
    print("  - monthly_returns_heatmap.png")
    print("  - drawdown_curves.png")
    print("  - returns_distribution.png")
    print("  - rolling_metrics.png")
    print("  - metric_comparison.png")
    print("  - feature_importance.png")
    print("  - prediction_scatter_*.png")
    print("  - performance_summary.csv")
    print("  - model_performance.csv")
    
    return {
        'results': results,
        'panel': panel2,
        'sym_arrays': sym_arrays,
        'xgb_model': xgb,
        'lstm_model': model,
        'performance_summary': perf_summary,
        'model_performance': model_perf,
        'feature_importance': feat_imp
    }
```

```python
# ============================================================================
# CELL 14: DETAILED REPORTING
# ============================================================================

def generate_detailed_report(pipeline_output):
    """Generate a comprehensive text report"""
    
    results = pipeline_output['results']
    
    report = []
    report.append("="*100)
    report.append("COMPREHENSIVE TRADING SYSTEM REPORT")
    report.append("="*100)
    report.append(f"\nGenerated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    report.append(f"\nConfiguration:")
    report.append(f"  Prediction Horizon: {CFG['horizon']} days")
    report.append(f"  Lookback Window: {CFG['lookback']} days")
    report.append(f"  Portfolio Size: Top {CFG['top_k']} stocks")
    report.append(f"  Rebalancing: Every {['Monday','Tuesday','Wednesday','Thursday','Friday'][CFG['rebalance_weekday']]}")
    report.append(f"  Transaction Costs: {CFG['cost_bps']} bps")
    report.append(f"  Slippage: {CFG['slip_bps']} bps")
    report.append(f"  Model Ensemble: {CFG['w_xgb']*100:.0f}% XGBoost + {CFG['w_lstm']*100:.0f}% LSTM")
    
    for period_name, res in results.items():
        if res is None:
            continue
        
        report.append(f"\n\n{'='*100}")
        report.append(f"{period_name.upper()} PERIOD RESULTS")
        report.append('='*100)
        
        stats = res['stats']
        
        report.append(f"\nPeriod: {stats['Start']} to {stats['End']}")
        report.append(f"\nStrategy Performance:")
        report.append(f"  Total Return        : {stats['Total_Return']:>10.2f}%")
        report.append(f"  CAGR                : {stats['CAGR']:>10.2f}%")
        report.append(f"  Sharpe Ratio        : {stats['Benchmark_Sharpe']:>10.2f}")
        report.append(f"  Max Drawdown        : {stats['Benchmark_MaxDD']:>10.2f}%")
        
        report.append(f"\nRelative Performance:")
        report.append(f"  Alpha               : {stats['Alpha']:>10.2f}%")
        report.append(f"  Outperformance      : {stats['Total_Return'] - stats['Benchmark_Return']:>10.2f}%")
        
        report.append(f"\nTrading Statistics:")
        report.append(f"  Number of Trades    : {stats['Num_Trades']:>10d}")
        report.append(f"  Avg Turnover        : {stats['Avg_Turnover']:>10.2f}%")
        report.append(f"  Total Costs         : {stats['Total_Cost']*100:>10.2f}%")
        report.append(f"  Win Rate            : {stats['Win_Rate']:>10.2f}%")
        report.append(f"  Best Day            : {stats['Best_Day']:>10.2f}%")
        report.append(f"  Worst Day           : {stats['Worst_Day']:>10.2f}%")
        
        # Monthly breakdown
        if len(res['daily_returns']) > 30:
            equity_series = pd.Series(res["equity"], index=pd.DatetimeIndex(res["dates"]))
            monthly = equity_series.resample('M').last().pct_change().dropna() * 100
            
            report.append(f"\nMonthly Returns:")
            for date, ret in monthly.items():
                report.append(f"  {date.strftime('%Y-%m')}: {ret:>8.2f}%")
    
    # Save report
    report_text = '\n'.join(report)
    with open('detailed_report.txt', 'w') as f:
        f.write(report_text)
    
    print(report_text)
    print(f"\n\nReport saved to: detailed_report.txt")
    
    return report_text

def export_trades_to_csv(results):
    """Export all trades to CSV for analysis"""
    all_trades = []
    
    for period_name, res in results.items():
        if res is None or not res.get('trades'):
            continue
        
        for trade in res['trades']:
            trade_copy = trade.copy()
            trade_copy['period'] = period_name
            trade_copy['holdings'] = ', '.join(trade_copy['holdings'])
            all_trades.append(trade_copy)
    
    if all_trades:
        df = pd.DataFrame(all_trades)
        df.to_csv('all_trades.csv', index=False)
        print(f"\nExported {len(all_trades)} trades to: all_trades.csv")
        return df
    
    return None
```

```python
# ============================================================================
# CELL 16: TRANSACTION REPORTING AND PROFIT IMPROVEMENT ANALYSIS
# ============================================================================

def export_detailed_transactions(results):
    """Export all buy/sell transactions to CSV with detailed breakdown"""
    all_transactions = []
    
    for period_name, res in results.items():
        if res is None or not res.get('detailed_transactions'):
            continue
        
        for trans in res['detailed_transactions']:
            trans_copy = trans.copy()
            trans_copy['period'] = period_name
            all_transactions.append(trans_copy)
    
    if all_transactions:
        df = pd.DataFrame(all_transactions)
        df = df.sort_values(['period', 'date', 'action'])
        df.to_csv('detailed_transactions.csv', index=False)
        print(f"\n Exported {len(all_transactions)} transactions to: detailed_transactions.csv")
        
        # Print summary by period
        print("\n" + "="*80)
        print(" TRANSACTION SUMMARY BY PERIOD")
        print("="*80)
        for period in df['period'].unique():
            period_data = df[df['period'] == period]
            buys = period_data[period_data['action'] == 'BUY']
            sells = period_data[period_data['action'] == 'SELL']
            print(f"\n{period}:")
            print(f"  Total Buys:  {len(buys):3d} transactions")
            print(f"  Total Sells: {len(sells):3d} transactions")
            print(f"  Buy Value:   {buys['value'].sum():,.2f}")
            print(f"  Sell Value:  {sells['value'].sum():,.2f}")
            print(f"  Net Flow:    {buys['value'].sum() - sells['value'].sum():,.2f}")
        
        # Most traded stocks
        print("\n" + "="*80)
        print(" MOST ACTIVELY TRADED STOCKS (ALL PERIODS)")
        print("="*80)
        stock_trades = df.groupby('symbol').size().sort_values(ascending=False).head(15)
        for stock, count in stock_trades.items():
            buys = len(df[(df['symbol'] == stock) & (df['action'] == 'BUY')])
            sells = len(df[(df['symbol'] == stock) & (df['action'] == 'SELL')])
            print(f"  {stock:20s} : {count:3d} trades (Buy: {buys:2d}, Sell: {sells:2d})")
        
        return df
    
    print("\n No transaction data found.")
    return None


def print_trade_log(results, period_name=None, max_trades=20):
    """Print detailed trade log with buy/sell breakdown"""
    if period_name:
        periods_to_show = {period_name: results.get(period_name)}
    else:
        periods_to_show = results
    
    for name, res in periods_to_show.items():
        if res is None or not res.get('trades'):
            continue
        
        print("\n" + "="*100)
        print(f" DETAILED TRADE LOG - {name}")
        print("="*100)
        
        for i, trade in enumerate(res['trades'][:max_trades]):
            equity_idx = min(i+1, len(res['equity'])-1)
            portfolio_value = res['equity'][equity_idx]
            
            print(f"\n{'='*100}")
            print(f"Trade #{i+1} - Date: {trade['date'].strftime('%Y-%m-%d (%A)')}")
            print(f"{'='*100}")
            print(f"Portfolio Value Before: {res['equity'][i]:,.4f}")
            print(f"Portfolio Value After:  {portfolio_value:,.4f}")
            print(f"Change:                 {portfolio_value - res['equity'][i]:,.4f}")
            
            print(f"\n BUYS ({len(trade['buys'])} stocks):")
            if trade['buys']:
                for stock in trade['buys']:
                    print(f"    {stock}")
            else:
                print("   (None)")
            
            print(f"\n SELLS ({len(trade['sells'])} stocks):")
            if trade['sells']:
                for stock in trade['sells']:
                    print(f"    {stock}")
            else:
                print("   (None)")
            
            print(f"\n HOLDS ({len(trade['holds'])} stocks):")
            if trade['holds']:
                for stock in trade['holds']:
                    print(f"    {stock}")
            else:
                print("   (None)")
            
            print(f"\n Transaction Costs:")
            print(f"   Turnover:     {trade['turnover']*100:6.2f}%")
            print(f"   Cost (bps):   {trade['cost']*10000:6.2f} basis points")
            print(f"   Cost ():     {trade['cost'] * portfolio_value:,.4f}")
        
        if len(res['trades']) > max_trades:
            print(f"\n{'='*100}")
            print(f"... and {len(res['trades']) - max_trades} more trades (use max_trades parameter to see more)")
        
        print("\n" + "="*100 + "\n")


def suggest_improvements(results, panel):
    """Analyze results and provide actionable improvement suggestions"""
    print("\n" + "="*100)
    print(" PROFIT IMPROVEMENT SUGGESTIONS")
    print("="*100)
    
    suggestions = []
    
    for period_name, res in results.items():
        if res is None:
            continue
        
        stats = res['stats']
        
        # Suggestion 1: High transaction costs
        if stats['Total_Cost'] * 100 > 2.0:
            suggestions.append({
                'period': period_name,
                'priority': ' HIGH',
                'issue': 'High Transaction Costs',
                'detail': f"Transaction costs are {stats['Total_Cost']*100:.2f}% of portfolio",
                'impact': f"Losing ~{stats['Total_Cost']*100:.1f}% to fees",
                'suggestion': "Reduce rebalancing from weekly to bi-weekly or monthly"
            })
        
        # Suggestion 2: High turnover
        if stats['Avg_Turnover'] > 0.5:
            suggestions.append({
                'period': period_name,
                'priority': ' MEDIUM',
                'issue': 'High Portfolio Turnover',
                'detail': f"Average turnover is {stats['Avg_Turnover']*100:.1f}%",
                'impact': "Excessive trading increases costs and taxes",
                'suggestion': "Add momentum persistence filter or widen selection threshold"
            })
        
        # Suggestion 3: Low Sharpe ratio
        if stats['Sharpe'] < 1.0:
            suggestions.append({
                'period': period_name,
                'priority': ' MEDIUM',
                'issue': 'Low Risk-Adjusted Returns',
                'detail': f"Sharpe ratio is {stats['Sharpe']:.2f} (target: >1.5)",
                'impact': "Returns not compensating for risk taken",
                'suggestion': "Add volatility filtering or implement risk parity weighting"
            })
        
        # Suggestion 4: Large drawdown
        if abs(stats['MaxDD']) > 25:
            suggestions.append({
                'period': period_name,
                'priority': ' HIGH',
                'issue': 'Excessive Drawdown',
                'detail': f"Maximum drawdown is {stats['MaxDD']:.1f}% (target: <20%)",
                'impact': "Large losses are psychologically difficult and take long to recover",
                'suggestion': "Implement 15% stop-loss or reduce position sizes during high volatility"
            })
        
        # Suggestion 5: Poor win rate
        if stats['Win_Rate'] < 50:
            suggestions.append({
                'period': period_name,
                'priority': ' MEDIUM',
                'issue': 'Low Win Rate',
                'detail': f"Win rate is only {stats['Win_Rate']:.1f}%",
                'impact': "More losing days than winning days",
                'suggestion': "Improve signal quality or adjust prediction horizon"
            })
        
        # Suggestion 6: Underperforming benchmark
        if stats['Alpha'] < 0:
            suggestions.append({
                'period': period_name,
                'priority': ' HIGH',
                'issue': 'Underperforming Benchmark',
                'detail': f"Alpha is {stats['Alpha']:.2f}% (negative)",
                'impact': "Strategy is losing to simple index investing",
                'suggestion': "Re-evaluate model features or consider passive indexing"
            })
    
    # Print period-specific suggestions
    if suggestions:
        for i, sug in enumerate(suggestions, 1):
            print(f"\n{i}. {sug['priority']} [{sug['period']}] {sug['issue']}")
            print(f"    Detail:     {sug['detail']}")
            print(f"    Impact:     {sug['impact']}")
            print(f"    Solution:   {sug['suggestion']}")
    else:
        print("\n Excellent! No major issues detected. Strategy is performing well!")
    
    # General improvement ideas
    print("\n" + "="*100)
    print(" GENERAL IMPROVEMENT IDEAS TO BOOST PROFITS")
    print("="*100)
    
    print("""
1  REDUCE COSTS (Immediate +1-2% profit):
    Change rebalancing from weekly to monthly
    Increase top_k from 10 to 15-20 stocks (less turnover per stock)
    Add 2% buffer zone before rebalancing (don't trade if change < 2%)
   
2  IMPROVE FEATURES (Potential +3-5% profit):
    Add sector momentum (relative performance vs sector average)
    Include market regime indicators (bull/bear/sideways detection)
    Add earnings announcement proximity
    Include institutional ownership changes
   
3  BETTER RISK MANAGEMENT (Reduce drawdowns 30-50%):
    Dynamic position sizing based on volatility (smaller size for volatile stocks)
    Stop-loss at -15% for individual positions
    Reduce exposure when VIX > 25
    Maximum 25% allocation to any single sector
   
4  MODEL IMPROVEMENTS (Potential +2-4% profit):
    Add LightGBM to ensemble (often outperforms XGBoost)
    Implement attention mechanism in LSTM
    Use meta-labeling to filter low-confidence predictions
    Add separate models for bull vs bear markets
   
5  PORTFOLIO OPTIMIZATION (Potential +2-3% profit):
    Replace equal-weighting with mean-variance optimization
    Use Kelly criterion for optimal position sizing
    Implement risk parity across sectors
    Add correlation matrix to avoid overcrowding similar stocks
   
6  ADVANCED TECHNIQUES (Potential +3-6% profit):
    Predict volatility separately and use for position sizing
    Add regime-switching models
    Use reinforcement learning for portfolio allocation
    Implement dynamic hedging with index options
   
7  DATA IMPROVEMENTS:
    Add alternative data (news sentiment, social media)
    Include insider trading data
    Add options flow data
    Monitor institutional holdings
    """)
    
    print("\n" + "="*100)
    print(" PROFIT CALCULATION EXAMPLE")
    print("="*100)
    
    # Calculate potential improvements
    for period_name, res in results.items():
        if res is None:
            continue
        
        stats = res['stats']
        print(f"\n{period_name}:")
        print(f"  Current Return:           {stats['Total_Return']:>8.2f}%")
        
        # Estimate improvements
        cost_saving = stats['Total_Cost'] * 50  # Save 50% of costs
        improved_sharpe = stats['Total_Return'] * 0.15  # 15% better returns
        lower_drawdown = abs(stats['MaxDD']) * 0.30  # 30% smaller drawdown = more capital
        
        total_improvement = cost_saving + improved_sharpe + lower_drawdown
        new_return = stats['Total_Return'] + total_improvement
        
        print(f"  + Cost Reduction:         {cost_saving:>8.2f}%")
        print(f"  + Better Features:        {improved_sharpe:>8.2f}%")
        print(f"  + Risk Management:        {lower_drawdown:>8.2f}%")
        print(f"  " + "-"*40)
        print(f"  Potential New Return:     {new_return:>8.2f}%")
        print(f"  Improvement:              +{total_improvement:>7.2f}%")
    
    print("\n" + "="*100)
    
    return suggestions


def print_final_profit_summary(results):
    """Print clean final profit summary"""
    print("\n" + "="*100)
    print(" FINAL PROFIT SUMMARY")
    print("="*100)
    
    for period_name, res in results.items():
        if res is None:
            continue
        
        stats = res['stats']
        initial_capital = 100000  # Assume 1,00,000 starting capital
        final_value = initial_capital * stats['Final_Portfolio_Value']
        profit = final_value - initial_capital
        
        print(f"\n{period_name} ({stats['Start']} to {stats['End']}):")
        print(f"  {'='*80}")
        print(f"  Starting Capital:     {initial_capital:>15,.2f}")
        print(f"  Ending Value:         {final_value:>15,.2f}")
        print(f"  Profit/Loss:          {profit:>15,.2f}")
        print(f"  Return:               {stats['Total_Return']:>15.2f}%")
        print(f"  CAGR:                 {stats['CAGR']:>15.2f}%")
        print(f"  Sharpe Ratio:         {stats['Sharpe']:>15.2f}")
        print(f"  Max Drawdown:         {stats['MaxDD']:>15.2f}%")
        print(f"  vs Benchmark:         {stats['Alpha']:>15.2f}% alpha")
        
        # Performance rating
        if stats['Sharpe'] > 2.0 and stats['Total_Return'] > 20:
            rating = " EXCELLENT"
        elif stats['Sharpe'] > 1.5 and stats['Total_Return'] > 15:
            rating = " VERY GOOD"
        elif stats['Sharpe'] > 1.0 and stats['Total_Return'] > 10:
            rating = " GOOD"
        elif stats['Total_Return'] > 0:
            rating = " FAIR"
        else:
            rating = " NEEDS IMPROVEMENT"
        
        print(f"\n  Performance Rating:   {rating}")
    
    print("\n" + "="*100)


# ============================================================================
# USAGE EXAMPLE - ADD THIS AT THE END OF YOUR NOTEBOOK
# ============================================================================

if __name__ == "__main__":
    # After running main_pipeline(), use these functions:
    
    print("\n" + "="*100)
    print("GENERATING ADDITIONAL REPORTS...")
    print("="*100)
    
    # 1. Export all transactions
    trans_df = export_detailed_transactions(pipeline_output['results'])
    
    # 2. Print detailed trade log (showing first 10 trades)
    print_trade_log(pipeline_output['results'], max_trades=10)
    
    # 3. Get improvement suggestions
    suggestions = suggest_improvements(pipeline_output['results'], pipeline_output['panel'])
    
    # 4. Print final profit summary
    print_final_profit_summary(pipeline_output['results'])
    
    print("\n" + "="*100)
    print(" ALL ANALYSIS COMPLETE!")
    print("="*100)
    print("\nGenerated files:")
    print("   detailed_transactions.csv  - All buy/sell transactions")
    print("   performance_summary.csv    - Performance metrics")
    print("   model_performance.csv      - Model accuracy")
    print("   all_trades.csv             - Trade history")
    print("   equity_curves.png          - Visual performance")
    print("   monthly_returns_heatmap.png")
    print("   drawdown_curves.png")
    print("   returns_distribution.png")
    print("   rolling_metrics.png")
    print("\n" + "="*100)
```
