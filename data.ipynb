```python
import yfinance as yf
import pandas as pd
import pandas_ta as ta
from pathlib import Path
```

```python
base = Path("data")
scalping_dir = base / "scalping"
positional_daily_dir = base / "positional" / "daily"
positional_hourly_dir = base / "positional" / "hourly"

scalping_dir.mkdir(parents=True, exist_ok=True)
positional_daily_dir.mkdir(parents=True, exist_ok=True)
positional_hourly_dir.mkdir(parents=True, exist_ok=True)
```

```python
tickers =["HDFCBANK.NS","ICICIBANK.NS","SBIN.NS","AXISBANK.NS"]
scalping_period = "60d"
scalping_interval = "5m"
positional_period= "2y"
dialy_interval = "1d"
hourly_interval = "1h"
```

```python
for ticker in tickers:
    
    df = yf.download(ticker,period=scalping_period,interval=scalping_interval,auto_adjust=False,progress=False)
    df= df.dropna(how="all")
    if not df.empty:
            df= df[["Open","High","Low","Close","Volume"]]
            scalp_file = scalping_dir / f"{ticker}_{scalping_interval}.csv"
            df.to_csv(scalp_file)
            print(f"SCALPING: {len(df)} rows saved to {scalp_file}")
    else:
        print("No scalping data")
    daily_df = yf.download(ticker,period=positional_period,interval=dialy_interval,auto_adjust=False,progress=False)
    daily_df = daily_df.dropna(how="all")
    if not daily_df.empty:
        daily_df = daily_df[["Open", "High", "Low", "Close", "Volume"]]
        daily_file = positional_daily_dir / f"{ticker}_daily.csv"
        daily_df.to_csv(daily_file)
        print(f" DAILY: {len(daily_df)} rows saved to {daily_file}")
    else:
        print("No dialy data")
    df_hourly = yf.download(ticker,period=positional_period,interval=hourly_interval,auto_adjust=False,progress=False)
    df_hourly = df_hourly.dropna(how="all")
    if not df_hourly.empty:
        df_hourly = df_hourly[["Open", "High", "Low", "Close", "Volume"]]
        hourly_file = positional_hourly_dir / f"{ticker}_hourly.csv"
        df_hourly.to_csv(hourly_file)
        print(f" HOURLY: {len(df_hourly)} rows saved to {hourly_file}")
    else:
        print(" No hourly data ")


   
 
    
  
```

```python
import time
from datetime import datetime
from datetime import time as dtime
ticker = "HDFCBANK.NS"
interval = "1m"
period = "1d"
sleep_time = 20
loop = 3 # for live fetching use infinte loop 


market_open =dtime(9,15)
market_close = dtime(15,30)
def is_open():
    now =datetime.now().time()
    return market_open<=now<=market_close
```

```python
def fetch(ticker,period,interval):
    df = yf.download(ticker,period=period,interval=interval,progress=False,auto_adjust=False)
    if df.empty:
        return None
    else:
        return df.tail(1)
for i in range(loop):
    print(f"Fetching at {datetime.now()}" )
    if not is_open():
        print("Market closed")
        break
    latest = fetch(ticker,period,interval)
    if latest is None:
        print("No data")
    else:
        latest = latest[["Open", "High", "Low", "Close", "Volume"]]
        print(latest)
    if i<loop-1:  
        print("sleeping for 20sec")
        time.sleep(sleep_time)
```
