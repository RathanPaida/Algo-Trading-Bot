```python
import warnings, math, numpy as np, pandas as pd
warnings.filterwarnings("ignore")

import yfinance as yf
from xgboost import XGBRegressor

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

import matplotlib.pyplot as plt
import logging
logging.getLogger("yfinance").setLevel(logging.CRITICAL)



CFG = {
    "start": "2010-01-01",
    "end":   "2025-02-01",
    "index": "^NSEI",

    "horizon": 10,
    "lookback": 60,

    "top_k": 10,
    "rebalance_weekday": 4,   #rebalance on friday

    "train_end": "2021-12-31",
    "val_end":   "2023-12-31",
    "test_start":"2024-01-01",
    "test_end":  "2024-12-31",

    "min_days": 850,

    "cost_bps": 10,
    "slip_bps": 5,

    "seed": 42,
    "device": "cuda" if torch.cuda.is_available() else "cpu",


    "lstm_epochs": 5,
    "lstm_batch": 256,
    "lstm_lr": 1e-3,
    "lstm_hidden": 48,
    "lstm_dropout": 0.2,

    "w_xgb": 0.8,
    "w_lstm": 0.2,

    "use_stock_gate": False,
}

np.random.seed(CFG["seed"])
torch.manual_seed(CFG["seed"])
torch.cuda.manual_seed_all(CFG["seed"])

SYMBOLS_2017 = ["ABB.NS", "ACC.NS", "AIAENG.NS", "ADANIENT.NS", "ADANIPORTS.NS", "ADANIPOWER.NS", "ABFRL.NS", "AJANTPHARM.NS", "ALKEM.NS", "AMARAJABAT.NS", "AMBUJACEM.NS", "APOLLOHOSP.NS", "APOLLOTYRE.NS", "ARVIND.NS", "ASHOKLEY.NS", "ASIANPAINT.NS", "AUROPHARMA.NS", "DMART.NS", "AXISBANK.NS", "BAJAJ-AUTO.NS", "BAJFINANCE.NS", "BAJAJFINSV.NS", "BALKRISIND.NS", "BANKBARODA.NS", "BANKINDIA.NS", "BATAINDIA.NS", "BERGEPAINT.NS", "BEL.NS", "BHARATFIN.NS", "BHARATFORG.NS", "BHEL.NS", "BPCL.NS", "BHARTIARTL.NS", "INFRATEL.NS", "BIOCON.NS", "BOSCHLTD.NS", "BRITANNIA.NS", "CESC.NS", "CRISIL.NS", "CADILAHC.NS", "CANBK.NS", "CASTROLIND.NS", "CENTRALBK.NS", "CENTURYTEX.NS", "CHOLAFIN.NS", "CIPLA.NS", "COALINDIA.NS", "COLPAL.NS", "CONCOR.NS", "COROMANDEL.NS", "CROMPTON.NS", "CUMMINSIND.NS", "DLF.NS", "DABUR.NS", "DALMIABHA.NS", "DHFL.NS", "DISHTV.NS", "DIVISLAB.NS", "LALPATHLAB.NS", "DRREDDY.NS", "EDELWEISS.NS", "EICHERMOT.NS", "EMAMILTD.NS", "ENDURANCE.NS", "ENGINERSIN.NS", "EXIDEIND.NS", "FEDERALBNK.NS", "FORTIS.NS", "GAIL.NS", "GMRINFRA.NS", "GSKCONS.NS", "GLAXO.NS", "GLENMARK.NS", "GODREJCP.NS", "GODREJIND.NS", "GRUH.NS", "GPPL.NS", "GSPL.NS", "HCLTECH.NS", "HDFCBANK.NS", "HAVELLS.NS", "HEROMOTOCO.NS", "HEXAWARE.NS", "HINDALCO.NS", "HINDPETRO.NS", "HINDUNILVR.NS", "HINDZINC.NS", "HUDCO.NS", "HDFC.NS", "ITC.NS", "ICICIBANK.NS", "ICICIPRULI.NS", "IDBI.NS", "IDFCBANK.NS", "IDFC.NS", "IRB.NS", "IDEA.NS", "IBULHSGFIN.NS", "INDIANB.NS", "INDHOTEL.NS", "IOC.NS", "IGL.NS", "INDUSINDBK.NS", "NAUKRI.NS", "INFY.NS", "INDIGO.NS", "IPCALAB.NS", "JSWENERGY.NS", "JSWSTEEL.NS", "JINDALSTEL.NS", "JUBLFOOD.NS", "JUBILANT.NS", "KARURVYSYA.NS", "KOTAKBANK.NS", "L&TFH.NS", "LICHSGFIN.NS", "LT.NS", "LUPIN.NS", "MRF.NS", "MGL.NS", "M&MFIN.NS", "M&M.NS", "MANAPPURAM.NS", "MRPL.NS", "MARICO.NS", "MARUTI.NS", "MFSL.NS", "MINDTREE.NS", "MOTHERSUMI.NS", "MPHASIS.NS", "MUTHOOTFIN.NS", "NATCOPHARM.NS", "NBCC.NS", "NHPC.NS", "NMDC.NS", "NTPC.NS", "NATIONALUM.NS", "OBEROIRLTY.NS", "ONGC.NS", "OIL.NS", "OFSS.NS", "PCJEWELLER.NS", "PIIND.NS", "PNBHOUSING.NS", "PAGEIND.NS", "PETRONET.NS", "PIDILITIND.NS", "PEL.NS", "PFC.NS", "POWERGRID.NS", "PRESTIGE.NS", "PGHH.NS", "PNB.NS", "RBLBANK.NS", "RAJESHEXPO.NS", "RCOM.NS", "RELIANCE.NS", "RELINFRA.NS", "RPOWER.NS", "RECLTD.NS", "SRF.NS", "SHREECEM.NS", "SRTRANSFIN.NS", "SIEMENS.NS", "SBIN.NS", "SAIL.NS", "STAR.NS", "SPARC.NS", "SUNPHARMA.NS", "SUNTV.NS", "SUZLON.NS", "SYNDIBANK.NS", "SYNGENE.NS", "TV18BRDCST.NS", "TVSMOTOR.NS", "TATACHEM.NS", "TATACOMM.NS", "TCS.NS", "TATAGLOBAL.NS", "TATAMTRDVR.NS", "TATAMOTORS.NS", "TATAPOWER.NS", "TATASTEEL.NS", "TECHM.NS", "RAMCOCEM.NS", "TITAN.NS", "TORNTPHARM.NS", "TORNTPOWER.NS", "UPL.NS", "ULTRACEMCO.NS", "UNIONBANK.NS", "UBL.NS", "MCDOWELL-N.NS", "VAKRANGEE.NS", "VEDL.NS", "VOLTAS.NS", "WELSPUNIND.NS", "WIPRO.NS", "WOCKPHARMA.NS", "YESBANK.NS", "ZEEL.NS"]

ALIASES = {
    "MOTHERSUMI.NS": ["MOTHERSON.NS"],
    "SRTRANSFIN.NS": ["SHRIRAMFIN.NS"],
    "TATAGLOBAL.NS": ["TATACONSUM.NS"],
    "INFRATEL.NS":   ["INDUSTOWER.NS"],
    "CADILAHC.NS":   ["ZYDUSLIFE.NS"],
    "TATAMTRDVR.NS": ["TATAMOTORS.NS"],
}
DROP_ALWAYS = set(["RCOM.NS", "DHFL.NS"])
SYMBOLS = [s for s in SYMBOLS_2017 if s not in DROP_ALWAYS]

def normalize_yf_columns(df):
    if df is None or df.empty: return df
    if isinstance(df.columns, pd.MultiIndex):
        df = df.copy()
        df.columns = df.columns.get_level_values(0)
    return df

def download_one(ticker):
    try:
        df = yf.download(ticker, start=CFG["start"], end=CFG["end"], progress=False, auto_adjust=False)
        df = normalize_yf_columns(df)
        if df is None or df.empty: return None
        need = ["Open","High","Low","Close","Volume"]
        if not all(c in df.columns for c in need): return None
        df = df[need].dropna()
        return None if df.empty else df
    except Exception:
        return None

def download_ohlcv(symbols, min_days):
    data, failed = {}, []
    for s in symbols:
        tried = [s] + ALIASES.get(s, [])
        ok = None
        used = None
        for t in tried:
            df = download_one(t)
            if df is not None and len(df) >= min_days:
                ok, used = df, t
                break
        if ok is None:
            failed.append(s)
        else:
            data[s] = ok
            if used != s:
                print(f"Alias used: {s} -> {used}")
    print(f"Download done. OK={len(data)} Failed={len(failed)}")
    if failed: print("Sample failed:", failed[:12])
    return data

def zscore(x, win=20):
    return (x - x.rolling(win).mean())/(x.rolling(win).std() + 1e-12)

def compute_features(df):
    out = pd.DataFrame(index=df.index)
    c,h,l = df["Close"], df["High"], df["Low"]
    v = df["Volume"].replace(0, np.nan).ffill()

    r1 = c.pct_change()
    out["ret1"] = r1
    out["ret5"] = c.pct_change(5)
    out["ret10"] = c.pct_change(10)
    out["mom20"] = c/c.shift(20) - 1

    out["vol20"] = r1.rolling(20).std()
    out["vol60"] = r1.rolling(60).std()

    out["sma10"] = c.rolling(10).mean()/c - 1
    out["sma20"] = c.rolling(20).mean()/c - 1
    out["sma50"] = c.rolling(50).mean()/c - 1
    out["sma200"] = c.rolling(200).mean()/c - 1

    tr = (h-l)/c
    out["tr"] = tr
    out["atr14"] = tr.rolling(14).mean()

    delta = c.diff()
    up = delta.clip(lower=0)
    down = (-delta).clip(lower=0)
    rs = up.rolling(14).mean()/(down.rolling(14).mean() + 1e-12)
    out["rsi14"] = 100 - 100/(1+rs)

    ema12 = c.ewm(span=12, adjust=False).mean()
    ema26 = c.ewm(span=26, adjust=False).mean()
    macd = ema12 - ema26
    sig = macd.ewm(span=9, adjust=False).mean()
    out["macd"] = macd/c
    out["macd_sig"] = sig/c
    out["macd_hist"] = (macd-sig)/c

    out["logvol_z20"] = zscore(np.log(v), 20)
    out["vol_chg5"] = v.pct_change(5)

    return out.replace([np.inf, -np.inf], np.nan)

FEATURES = ["ret1","ret5","ret10","mom20","vol20","vol60","sma10","sma20","sma50","sma200",
            "tr","atr14","rsi14","macd","macd_sig","macd_hist","logvol_z20","vol_chg5"]

def build_symbol_arrays(data_dict):
    sym_arrays={}
    panel_rows=[]

    for sym, df in data_dict.items():
        feat = compute_features(df)[FEATURES]
        y = (df["Close"].shift(-CFG["horizon"])/df["Close"] - 1.0).astype(float)

        tmp = feat.copy()
        tmp["y"] = y
        tmp["close"] = df["Close"]
        tmp = tmp.dropna(subset=FEATURES + ["y","close"])
        if len(tmp) < (CFG["lookback"] + 300):
            continue

        dates = pd.to_datetime(tmp.index)
        X = tmp[FEATURES].values.astype(np.float32)
        yv = tmp["y"].values.astype(np.float32)
        close = tmp["close"].values.astype(np.float32)

        sym_arrays[sym] = {"dates": dates, "X": X, "y": yv, "close": close}

        part = pd.DataFrame(X, columns=FEATURES)
        part["y"] = yv
        part["symbol"] = sym
        part["date"] = dates
        panel_rows.append(part)

    panel = pd.concat(panel_rows, ignore_index=True)
    panel = panel.sort_values(["date","symbol"]).reset_index(drop=True)
    return sym_arrays, panel

def make_pairs(sym_arrays):
    pairs=[]
    for sym, d in sym_arrays.items():
        for t in range(CFG["lookback"], len(d["dates"])):
            pairs.append((sym,t))
    return pairs


class SeqSliceDataset(Dataset):
    def __init__(self, sym_arrays, pairs, start_date=None, end_date=None, standardize=None):
        self.sym_arrays=sym_arrays
        sd = pd.to_datetime(start_date) if start_date else None
        ed = pd.to_datetime(end_date) if end_date else None

        kept=[]
        for sym,t in pairs:
            d = sym_arrays[sym]["dates"][t]
            if sd is not None and d < sd: continue
            if ed is not None and d > ed: continue
            kept.append((sym,t))
        self.pairs=kept
        self.mu, self.sig = (standardize if standardize else (None,None))

    def __len__(self): return len(self.pairs)

    def __getitem__(self,i):
        sym,t = self.pairs[i]
        arr = self.sym_arrays[sym]
        X = arr["X"][t-CFG["lookback"]:t]
        y = arr["y"][t]
        if self.mu is not None:
            X = (X - self.mu)/self.sig
        date_int = int(arr["dates"][t].value)
        return torch.from_numpy(X), torch.tensor([y],dtype=torch.float32), sym, torch.tensor(date_int,dtype=torch.int64)

class LSTMRegressor(nn.Module):
    def __init__(self, n_features, hidden=48, dropout=0.2):
        super().__init__()
        self.ln = nn.LayerNorm(n_features)
        self.lstm = nn.LSTM(n_features, hidden, batch_first=True)
        self.drop = nn.Dropout(dropout)
        self.fc = nn.Sequential(nn.Linear(hidden, 32), nn.ReLU(), nn.Linear(32, 1))
    def forward(self, x):
        x = self.ln(x)
        x,_ = self.lstm(x)
        x = self.drop(x)
        return self.fc(x[:, -1, :])

def train_lstm(train_loader, val_loader, n_features):
    device = CFG["device"]
    model = LSTMRegressor(n_features, hidden=CFG["lstm_hidden"], dropout=CFG["lstm_dropout"]).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=CFG["lstm_lr"])
    loss_fn = nn.MSELoss()

    best=float("inf"); best_state=None
    patience=2; bad=0

    for ep in range(1, CFG["lstm_epochs"]+1):
        model.train()
        tl=[]
        for xb,yb,_,_ in train_loader:
            xb,yb = xb.to(device), yb.to(device)
            opt.zero_grad(set_to_none=True)
            pred=model(xb)
            loss=loss_fn(pred,yb)
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            opt.step()
            tl.append(loss.item())

        model.eval()
        vl=[]
        with torch.no_grad():
            for xb,yb,_,_ in val_loader:
                xb,yb = xb.to(device), yb.to(device)
                vl.append(loss_fn(model(xb), yb).item())
        v=float(np.mean(vl)) if vl else float("inf")
        print(f"LSTM Epoch {ep:02d} train_mse={np.mean(tl):.6f} val_mse={v:.6f}")

        if v < best - 1e-6:
            best=v
            best_state={k:vv.detach().cpu().clone() for k,vv in model.state_dict().items()}
            bad=0
        else:
            bad += 1
            if bad >= patience:
                print("Early stop")
                break

    if best_state is not None:
        model.load_state_dict(best_state)
    return model

@torch.no_grad()
def predict_lstm(model, loader):
    device = CFG["device"]
    model.eval()
    out=[]
    for xb, _, syms, date_int in loader:
        xb = xb.to(device)
        pred = model(xb).detach().cpu().numpy().reshape(-1)
        dates = pd.to_datetime(date_int.numpy(), unit="ns")
        out.extend(list(zip(dates, list(syms), pred)))
    df = pd.DataFrame(out, columns=["date","symbol","pred_lstm"])
    df["date"] = pd.to_datetime(df["date"])
    return df


def sharpe(returns, ann=252):
    r=np.asarray(returns,float)
    return 0.0 if len(r)<2 or r.std()==0 else (r.mean()/r.std())*math.sqrt(ann)

def max_drawdown(equity):
    e=np.asarray(equity,float)
    peak=np.maximum.accumulate(e)
    dd=(e/peak)-1
    return float(dd.min())

def cagr(equity, ann=252):
    e=np.asarray(equity,float)
    if len(e)<2: return 0.0
    years=(len(e)-1)/ann
    return 0.0 if years<=0 else float((e[-1]/e[0])**(1/years)-1)

def get_rebalance_dates(dates, weekday=4):
    idx = pd.DatetimeIndex(sorted(pd.to_datetime(dates).unique()))
    return idx[idx.weekday == weekday]

def run_backtest(panel_pred, sym_arrays, idx_close):
    panel_pred = panel_pred.copy()
    panel_pred["date"] = pd.to_datetime(panel_pred["date"])

    start = pd.to_datetime(CFG["test_start"])
    end   = pd.to_datetime(CFG["test_end"])

    dates = pd.DatetimeIndex(sorted(panel_pred["date"].unique()))
    dates = dates[(dates >= start) & (dates <= end)]
    rebal = set(get_rebalance_dates(dates, CFG["rebalance_weekday"]))

    equity=[1.0]; eq_dates=[dates[0]]; daily=[]
    pos={}

    for i in range(1,len(dates)):
        d=dates[i]; prev=dates[i-1]

        day=0.0
        for sym,w in pos.items():
            arr=sym_arrays.get(sym)
            if arr is None: continue
            dt=arr["dates"].values
            t0=np.searchsorted(dt, np.datetime64(prev))
            t1=np.searchsorted(dt, np.datetime64(d))
            if t0>=len(arr["close"]) or t1>=len(arr["close"]): continue
            if dt[t0]!=np.datetime64(prev) or dt[t1]!=np.datetime64(d): continue
            day += w * float(arr["close"][t1]/arr["close"][t0]-1.0)

        equity.append(equity[-1]*(1+day))
        eq_dates.append(d)
        daily.append(day)

        if d not in rebal:
            continue

        today = panel_pred[panel_pred["date"]==d].dropna(subset=["score"]).sort_values("score", ascending=False)
        if len(today) < CFG["top_k"]:
            continue

        picks = today["symbol"].head(CFG["top_k"]).tolist()
        w = 1.0/len(picks)
        new_pos = {s:w for s in picks}

        union=set(pos)|set(new_pos)
        turnover=sum(abs(new_pos.get(s,0.0)-pos.get(s,0.0)) for s in union)
        cost=(CFG["cost_bps"]+CFG["slip_bps"])*1e-4*turnover
        equity[-1]*=(1-cost)

        pos=new_pos

    idx_close = idx_close.dropna()
    idx_ret = idx_close.reindex(eq_dates).pct_change().fillna(0.0).values
    idx_eq=[1.0]
    for r in idx_ret[1:]:
        idx_eq.append(idx_eq[-1]*(1+float(r)))

    stats={
        "CAGR": cagr(equity),
        "Sharpe": sharpe(daily),
        "MaxDD": max_drawdown(equity),
        "CAGR_NIFTY": cagr(idx_eq),
        "Sharpe_NIFTY": sharpe(idx_ret[1:]),
        "MaxDD_NIFTY": max_drawdown(idx_eq),
    }
    return {"dates": eq_dates, "equity": equity, "idx_equity": idx_eq, "stats": stats}


print("Device:", CFG["device"])

print("Downloading index...")
idx_df = yf.download(CFG["index"], start=CFG["start"], end=CFG["end"], progress=False, auto_adjust=False)
idx_df = normalize_yf_columns(idx_df)
idx_close = idx_df["Close"].dropna()
idx_close.index = pd.to_datetime(idx_close.index)

print("Downloading stocks...")
data = download_ohlcv(SYMBOLS, CFG["min_days"])
if len(data) < 30:
    raise RuntimeError("Too few symbols downloaded. Lower min_days to 650 and rerun.")

print("Building arrays + panel...")
sym_arrays, panel = build_symbol_arrays(data)
print("Symbols kept:", len(sym_arrays), "| panel rows:", len(panel))


train_end = pd.to_datetime(CFG["train_end"])
tr = panel[panel["date"] <= train_end].dropna(subset=FEATURES + ["y"])
print("Training XGBoost on", len(tr), "rows...")
xgb = XGBRegressor(
    n_estimators=1200, learning_rate=0.03, max_depth=6,
    subsample=0.8, colsample_bytree=0.8, reg_lambda=2.0,
    objective="reg:squarederror", random_state=CFG["seed"], n_jobs=4
)
xgb.fit(tr[FEATURES].values, tr["y"].values)
panel["pred_xgb"] = xgb.predict(panel[FEATURES].values)

trX = panel.loc[panel["date"] <= train_end, FEATURES].values.astype(np.float32)
mu = trX.mean(axis=0, keepdims=True)
sig = trX.std(axis=0, keepdims=True) + 1e-6

pairs = make_pairs(sym_arrays)

train_ds = SeqSliceDataset(sym_arrays, pairs, start_date="2010-01-01", end_date=CFG["train_end"], standardize=(mu, sig))
val_ds   = SeqSliceDataset(sym_arrays, pairs, start_date="2022-01-01", end_date=CFG["val_end"], standardize=(mu, sig))

train_loader = DataLoader(train_ds, batch_size=CFG["lstm_batch"], shuffle=True, num_workers=0, drop_last=True)
val_loader   = DataLoader(val_ds, batch_size=CFG["lstm_batch"], shuffle=False, num_workers=0)

print("LSTM samples train:", len(train_ds), "val:", len(val_ds))
model = train_lstm(train_loader, val_loader, n_features=len(FEATURES))

pred_ds = SeqSliceDataset(sym_arrays, pairs, start_date="2010-01-01", end_date=CFG["test_end"], standardize=(mu, sig))
pred_loader = DataLoader(pred_ds, batch_size=1024, shuffle=False, num_workers=0)
pred_lstm_df = predict_lstm(model, pred_loader)

panel2 = panel.merge(pred_lstm_df, on=["date","symbol"], how="left")
panel2["pred_lstm"] = panel2["pred_lstm"].fillna(panel2["pred_xgb"])
panel2["score"] = CFG["w_xgb"]*panel2["pred_xgb"] + CFG["w_lstm"]*panel2["pred_lstm"]

res = run_backtest(panel2, sym_arrays, idx_close)

print("\n 2024 backtesting: ")
for k,v in res["stats"].items():
    print(f"{k:12s}: {v:.4f}")

plt.figure(figsize=(10,5))
plt.plot(res["dates"], res["equity"], label="Strategy (XGB+LSTM)")
plt.plot(res["dates"], res["idx_equity"], label="NIFTY")
plt.title("Equity Curve(2024)")
plt.legend()
plt.grid(True)
plt.show()

eq = pd.Series(res["equity"], index=pd.DatetimeIndex(res["dates"]))
mret = eq.resample("M").last().pct_change().dropna()
print("\nMonthly returns (2024):")
display(mret.to_frame("strategy_monthly_return"))
```
