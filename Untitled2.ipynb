```python
import yfinance as yf
import pandas as pd
import pandas_ta as ta
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# --- PORTFOLIO CONFIG ---
# Major Indian Banking Stocks
SYMBOLS = ["HDFCBANK.NS", "ICICIBANK.NS", "SBIN.NS", "KOTAKBANK.NS", "AXISBANK.NS", "AUBANK.NS"]
CAPITAL_PER_STOCK = 100000  # Allocating 1 Lakh per stock
RISK_PER_TRADE = 0.01       # 1% risk
KILL_SWITCH_LIMIT = 0.05    # 5% Portfolio level stop

all_trade_logs = []
print(f"Portfolio Bot initialized for {len(SYMBOLS)} stocks.")
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning) # Hide the warnings

def get_processed_data(symbol):
    try:
        # Fixed the auto_adjust=True to remove warnings
        d_data = yf.download(symbol, period="2y", interval="1d", progress=False, auto_adjust=True)
        h_data = yf.download(symbol, period="2y", interval="1h", progress=False, auto_adjust=True)
        
        if d_data.empty or h_data.empty: return None, None
        if isinstance(d_data.columns, pd.MultiIndex): d_data.columns = d_data.columns.get_level_values(0)
        if isinstance(h_data.columns, pd.MultiIndex): h_data.columns = h_data.columns.get_level_values(0)

        # --- DAILY INDICATORS ---
        d_data['EMA_200'] = ta.ema(d_data['Close'], length=200)
        # Calculate Slope: Is the EMA 200 higher than it was 5 days ago?
        d_data['EMA_Slope'] = d_data['EMA_200'].diff(5) 
        d_data['RSI'] = ta.rsi(d_data['Close'], length=14)
        d_data['ATR'] = ta.atr(d_data['High'], d_data['Low'], d_data['Close'], length=14)
        adx_df = ta.adx(d_data['High'], d_data['Low'], d_data['Close'], length=14)
        d_data['ADX'] = adx_df['ADX_14']

        # --- HOURLY INDICATORS ---
        h_data['EMA_9'] = ta.ema(h_data['Close'], length=9)
        h_data['EMA_21'] = ta.ema(h_data['Close'], length=21)
        h_data['Vol_Avg'] = h_data['Volume'].rolling(20).mean()
        
        return d_data.dropna(), h_data.dropna()
    except Exception as e:
        print(f"Error fetching {symbol}: {e}")
        return None, None

def get_expiry_t2_date(date_obj):
    next_month = date_obj.replace(day=28) + timedelta(days=4)
    last_day = next_month - timedelta(days=next_month.day)
    offset = (last_day.weekday() - 3) % 7
    last_thursday = last_day - timedelta(days=offset)
    return (last_thursday - timedelta(days=2)).date()
```

```python
all_trade_logs = []

for symbol in SYMBOLS:
    daily_df, hourly_df = get_processed_data(symbol)
    if daily_df is None: continue

    current_pos = None
    entry_price = 0
    stop_loss = 0
    target_price = 0

    for i in range(2, len(hourly_df)):
        now = hourly_df.index[i]
        price = hourly_df['Close'].iloc[i]
        
        # Get Daily Context
        prev_day = now.date() - timedelta(days=1)
        day_context = daily_df.loc[daily_df.index.date <= prev_day]
        if day_context.empty: continue
        day_data = day_context.iloc[-1]

        # EXIT LOGIC
        if current_pos:
            if (current_pos == 'LONG' and (price <= stop_loss or price >= target_price)) or \
               (current_pos == 'SHORT' and (price >= stop_loss or price <= target_price)):
                
                pnl = (price - entry_price) if current_pos == 'LONG' else (entry_price - price)
                qty = (CAPITAL_PER_STOCK * RISK_PER_TRADE) / abs(entry_price - stop_loss)
                all_trade_logs.append({'symbol': symbol, 'time': now, 'pnl_cash': pnl * qty})
                current_pos = None
            continue

        # --- STRONGER ENTRY FILTERS ---
        # 1. Trend Strength (ADX > 20) and EMA Slope (Trend Direction)
        is_bullish_trend = (day_data['Close'] > day_data['EMA_200']) and (day_data['EMA_Slope'] > 0)
        is_bearish_trend = (day_data['Close'] < day_data['EMA_200']) and (day_data['EMA_Slope'] < 0)
        
        strong_momentum = day_data['ADX'] > 20

        if not current_pos and strong_momentum:
            # BUY LOGIC
            if is_bullish_trend and (50 < day_data['RSI'] < 65):
                # 9 EMA crosses above 21 EMA
                if hourly_df['EMA_9'].iloc[i] > hourly_df['EMA_21'].iloc[i] and \
                   hourly_df['EMA_9'].iloc[i-1] <= hourly_df['EMA_21'].iloc[i-1]:
                    
                    # Pullback check: Price shouldn't be more than 2% away from EMA 21
                    if price < (hourly_df['EMA_21'].iloc[i] * 1.02):
                        current_pos = 'LONG'
                        entry_price = price
                        sl_dist = 1.5 * day_data['ATR'] # Tighter SL to improve RR
                        stop_loss = entry_price - sl_dist
                        target_price = entry_price + (sl_dist * 3) # Target 1:3

            # SELL LOGIC
            elif is_bearish_trend and (35 < day_data['RSI'] < 50):
                if hourly_df['EMA_9'].iloc[i] < hourly_df['EMA_21'].iloc[i] and \
                   hourly_df['EMA_9'].iloc[i-1] >= hourly_df['EMA_21'].iloc[i-1]:
                    
                    if price > (hourly_df['EMA_21'].iloc[i] * 0.98):
                        current_pos = 'SHORT'
                        entry_price = price
                        sl_dist = 1.5 * day_data['ATR']
                        stop_loss = entry_price + sl_dist
                        target_price = entry_price - (sl_dist * 3)

portfolio_results = pd.DataFrame(all_trade_logs)
```

```python
if not portfolio_results.empty:
    # Aggregating results
    portfolio_results = portfolio_results.sort_values('time')
    portfolio_results['cum_pnl'] = portfolio_results['pnl_cash'].cumsum()
    
    # Calculate Metrics
    total_profit = portfolio_results['pnl_cash'].sum()
    win_rate = (len(portfolio_results[portfolio_results['pnl_cash'] > 0]) / len(portfolio_results)) * 100
    max_drawdown = (portfolio_results['cum_pnl'].cummax() - portfolio_results['cum_pnl']).max()
    
    print("-" * 30)
    print(f"PORTFOLIO ANALYSIS")
    print("-" * 30)
    print(f"Total Net Profit: ₹{total_profit:.2f}")
    print(f"Win Rate: {win_rate:.2f}%")
    print(f"Max Portfolio Drawdown: ₹{max_drawdown:.2f}")
    print(f"Total Trades Executed: {len(portfolio_results)}")
    print("-" * 30)

    # Plot Equity Curve
    plt.figure(figsize=(12, 6))
    plt.plot(portfolio_results['time'], portfolio_results['cum_pnl'], color='green', linewidth=2)
    plt.fill_between(portfolio_results['time'], portfolio_results['cum_pnl'], alpha=0.2, color='green')
    plt.title("Portfolio Equity Curve (Combined Banking Stocks)")
    plt.xlabel("Date")
    plt.ylabel("Cumulative Profit (₹)")
    plt.grid(True, alpha=0.3)
    plt.show()
else:
    print("No trades found for the selected symbols and criteria.")
    
```
